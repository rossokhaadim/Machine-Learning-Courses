13. Write a function that would multiply all values in a column that are less than the average value in this column by 2, and divide the other values by 4



```python
def calc(array, i):
  mean = np.mean(array, axis = 0)
  result = np.where(array < mean, array * 2, array / 4).transpose()[i]
  return result
```

Apply the obtained function to the 3rd column

```python
calc(new_iris, 2)
```
